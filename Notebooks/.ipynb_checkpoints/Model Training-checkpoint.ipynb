```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("C:/Users/keshv/OneDrive/Desktop/Data_Science/Jupyter lab/07 Machine learning/Clean Vehicle.csv")
```

```python
X = df.drop(columns = "price")
y = df["price"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .25, shuffle = True, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
Scaler = StandardScaler()
X_train = Scaler.fit_transform(X_train)
X_test = Scaler.transform(X_test)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
linear = linear.fit(X_train, y_train)
y_pred = linear.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
score = r2_score(y_test, y_pred)
Adjusted_score = 1 - (1-score) * (len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```

```python
print("mse =", mse)
print("mae =", mae)
print("rmse =", rmse)
print("r2 =", score)
print("Adjusted r2 =", Adjusted_score)
```

## Lasso CV

```python
from sklearn.linear_model import LassoCV
lassocv = LassoCV(alphas = [.01, .1, 1, 10, 100])
```

```python
print("mse =", mse)
print("mae =", mae)
print("rmse =", rmse)
print("r2 =", score)
print("Adjusted r2 =", Adjusted_score)
```

## ElasticNet CV

```python
from sklearn.linear_model import ElasticNetCV
elnetcv = ElasticNetCV(alphas = [.01, .1, 1, 10, 100], l1_ratio = [.2, .5, .8])
elnetcv = elnetcv.fit(X_train, y_train)
```

```python
y_pred = elnetcv.predict(X_test)
```
