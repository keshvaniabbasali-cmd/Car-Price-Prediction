```python
df = pd.get_dummies(df, columns = ["manufacturer", "type", "condition", "paint_color"], dtype = int)
```

```python
df.drop(columns = ["model", "region", "posting_date"], inplace = True)
```

```python
print(df["price"].describe())
print((df["price"] == 0).sum())
print((df["price"] < 0).sum())
print(df["price"].max())
```

```python
Q1 = df["price"].quantile(0.25)
Q3 = df["price"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df["price"] >= lower) & (df["price"] <= upper)]
```

```python
X = df.drop(columns = "price")
y = df["price"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .25, random_state = 42, shuffle = True)
```

```python
from sklearn.preprocessing import StandardScaler
Scaler = StandardScaler()
X_train = Scaler.fit_transform(X_train)
X_test = Scaler.transform(X_test)
```

## LinearRegression

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg = linreg.fit(X_train, y_train)
y_pred = linreg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
score = r2_score(y_test, y_pred)
Adjusted_score = 1 - (1-score) * (len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```

```python
print("mse =", mse)
print("mae =", mae)
print("rmse =", rmse)
print("r2 =", score)
print("Adjusted r2 =", Adjusted_score)
```

## Ridge

```python
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge = ridge.fit(X_train, y_train)
y_pred = ridge.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
score = r2_score(y_test, y_pred)
Adjusted_score = 1 - (1-score) * (len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```

```python
print("mse =", mse)
print("mae =", mae)
print("rmse =", rmse)
print("r2 =", score)
print("Adjusted r2 =", Adjusted_score)
```
